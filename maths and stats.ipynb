```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import csv
```

```python
# upload dataset
#first convert dataset into csv format and then as file is encoded in latin-1,we have to use encoding.
df = pd.read_csv('sample-superstore-subset-excel.csv', encoding='latin-1')
```

```python
#display first few rows
df.head()
```

```python
#summary statistics
df.describe()
```

```python
# we have to check for missing values
df.isnull().sum()
```

```python
# As we can see there are 72 missing values in Product Base Margin column.because this is numerical column we will fill missing values with mean.
df["Product Base Margin"]=df["Product Base Margin"].fillna(df["Product Base Margin"].mean())
```

```python
# check for datatypes
df.dtypes
```

```python
# As we can see 'order date' and 'ship date' are in object format, we have to change it into datetime format.
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
```

```python
# creating new columns

df["Gross profit"] = df["Profit"]/df["Sales"]
```

```python
#  univariate and Bivariate analysis 
# Histogram for sales column
plt.figure(figsize=(10,5))
sns.histplot(df["Sales"],color="blue",alpha=0.7,kde=True)
plt.xlabel("Sales")
plt.ylabel("count")
plt.title("Distribution by Sales")

# countplot for orders by region
plt.figure(figsize=(10,5))
sns.countplot(data=df,x="Region",color="green")
plt.xlabel("Region")
plt.ylabel("count")
plt.title("count of orders by Region")

# Scatter plot for sales vs profit
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x="Sales",y="Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vc Profit")


# Boxplot for sales by region
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x="Region",y="Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.title("sales by region")
```

```python
# Summary statistics for key variables
# Key variables
key_variable = ["Sales","Profit","Quantity ordered new","Discount"]

# now  summary stats
stats=df[key_variable].describe()
print(stats)

# find profit for each product category 
pc_profit=df.groupby("Product Category")["Profit"].sum()

print(pc_profit)

# now find sales in different Regions
RegionWiseSales = df.groupby("Region")["Sales"].sum()

print(RegionWiseSales)


# from above insights we xan conclude that Technology having hieghst profit where as Furniture having lowest and sale is more in Central region where as less in South.
```

```python
from scipy import stats
from scipy.stats import ttest_ind, chi2_contingency

# T-test for sales in different regions
# from above results we having Resion wise sales
Central = 2540341.62
East = 2422804.68
South = 1597346.22
West = 2391438.80

# t-test
t_statastics , p_value = stats.ttest_ind(Central,East,equal_var=False)

print(t_statastics , p_value )
```

```python
# Chi-square test for independence between Ship Mode and Customer Segment

from scipy.stats import ttest_ind, chi2_contingency
# define Ho and Ha 
Ha = "There is relationship between Ship Mode and Customer Segment"
Ho = "There is no relationship between Ship Mode and Customer Segment"

# create contigency table
contigency_table = pd.crosstab(df["Ship Mode"],df["Customer Segment"])

# calculate p-value and ch2
p_value  = chi2_contingency(contigency_table)
chi2  = chi2_contingency(contigency_table)
# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p_value)
# results
alpha = 0.05
if 'p_value' < 'alpha':
    print(Ha)
else:
    print(Ho)
```

```python
# Identifying top-performing product categories based on total sales and profit

category = df.groupby('Product Category').aggregate({"Sales":"sum","Profit":"sum"})

#now top category for sales
Top_category_sales = category.sort_values("Sales",ascending=False)

# top category for profit
Top_category_Profit = category.sort_values("Profit",ascending=False)

# now we want top 10
print("top 10 :\n",Top_category_sales.head(10))

```

```python
# Plotting total sales by product category
plt.figure(figsize=(10,5))
sns.barplot(x=Top_category_sales.index, y=Top_category_sales["Sales"],palette='viridis')
plt.xlabel("product category")
plt.ylabel("total sales")
plt.title("Total sales by product category")
    
    
plt.figure(figsize=(10,5))
sns.barplot(x=Top_category_Profit.index, y=Top_category_Profit["Profit"],palette='viridis')
plt.xlabel("product category")
plt.ylabel("total profit")
plt.title("Total profit by product category")
```
